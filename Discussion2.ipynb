In this video I think it is important to see the way a true hybrid recommender works and the ways it can come into issues in industry. The first things that interested me were the use of matrix factorization to create patterns that can be learned from because as a mmusic listener I rarely rate anything. This missing data ould be a point of contention if the ratings were a heavy use for the system. This gets even stronger by using the Alternating Least Squares method to find a point of convergence by alternating the user and item vectors. There's also a great value here in using a binary system of plays for each song. One rating may be reflective of a time but a matrix showing a heavy distribution of plays to a certain item are extremely telling for a user's tendencies.

The second portion of this video was even more interesting to me because I feel implementation is everything. Playing a song many times is valuable but at a certain scale, we can't just endlessly iterate for every user or objectively ignore resource allocation in our creation of systems. Obviously we would rather create a good system and solve for this later, but there must be some accounting for this. A system like spark that can use a data cache in memory as opposed to Hadoop. Clearly spark posed other challenges as dicsussed here for some things like memory limits, network shuffling data between machines in ALS, and partitioning best for the use of the user and item vectors. If this requires even more shuffling, all processes may can be slowed or interrupted. The combination of machine learning, distributed computing, and data management has been a big question in my mind so far, and this video is an excellent example of how these problems are addressed at an industry level.
