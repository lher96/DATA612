I personally believe that there needs to be a 3rd party organization that monitors recommender systems used by social media. Adding a governing body can help create a standard for how these systems are meant to mitigate the spread of dangerous content.

In the WIRED article its discussed how models that aim for engagement can be easily prone towards pushing misinformation, conspiracy theories, extremism, and political polarization. This is all perfectly exemplified in the author's example of her Pinterest board causing her to be fed far-right and Russian content. This is to be expected as collaborative and content based filtering both aim to push content that fills their parameters but do not police the content itself. This is where I think she goes into a very valuable idea of a recommender system. There should be "ethical recommendation" systems that are filtering content better. Content with shock value will typically produce high engagement but amplifying ideas based on this engagement not only devalues the platform but the system itself once rolling will create a cascade of poor recommendations.

The berkeley paper makes some interesting points adjacent to the WIRED article about social influence bias. User ratings and engagement can be seriously influenced by the community. The number of likes, comments, sentiment amongst comments, can all very well contribute to how someone views content. When a bot farm can comment pro-fascist or pro-violent comments on posts, it can create normalization for people who believe this idea is prevelant and accepted. Over time we have seen how this has the power to influence communities and convince users of misinformation and conspiracy. In this paper they apply a machine learning algorithm that uses pre and post bias ratings to create a bias mitigation process. I believe this may be a better way of dealing with this issue than trying to give an external or internal entity the right to police ethical standards of content. That can be a slippery slope and severely alter the data for the recommender and we would clearly want to avoid sacrificing too much accuracy for ethics. It begs the interesting question though, are we separating bias and user conditioning from accuracy and understanding of our model's system and processes. I would love for this to be considered more in the greater regulation of social media but the denial of experts that has grown through social media is an interesting obstacle in this.
